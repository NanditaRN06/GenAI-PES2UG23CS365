```python
from transformers import pipeline

import warnings
warnings.filterwarnings('ignore')
```

```python
translator_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
translator_es = pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")
```

**Hardcoded input**

```python
text = "This user manual explains how to install the software safely."

result_fr = translator_fr(text)
print("French:", result_fr[0]["translation_text"])

result_es = translator_es(text)
print("Spanish:", result_es[0]["translation_text"])
```

**User input**

```python
print("-- Language Translation Assistant --")
print("Type 'quit()' anytime to stop\n")

while True:
    choice = input("Choose language (1 for French / 2 for Spanish): ").strip()

    if choice.lower() == "quit()":
        print("Exiting translator. Bye ...")
        break

    if choice not in ["1", "2"]:
        print("Invalid choice. Please enter 1 or 2.\n")
        continue

    text = input("Enter English text: ")

    if text.strip().lower() == "quit()":
        print("Exiting translator. Bye ...")
        break

    if text.strip() == "":
        print("Please enter some text.\n")
        continue

    if choice == "1":
        result = translator_fr(text)
        print("French Translation:", result[0]["translation_text"], "\n")

    elif choice == "2":
        result = translator_es(text)
        print("Spanish Translation:", result[0]["translation_text"], "\n")
```

**Observation:** The notebook successfully translated the given English sentences into French and Spanish using Hugging Face translation pipelines, and the outputs were clear, grammatically correct, and contextually accurate for both hardcoded inputs and user-provided text.
